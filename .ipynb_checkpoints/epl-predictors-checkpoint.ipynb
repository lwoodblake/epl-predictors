# epl-predictors

Expected goals are a football (soccer) metric used to define the probability that a shot taken will result in a goal. A shot with an xG of 0.5 will result in a goal half of the time. Each goalscoring opportunity has an xG value. xG is accumulated over the course of a game and over the course of a season. Individual players accumulate xG, as well as clubs as a whole. More can be read about xG [here](https://fbref.com/en/expected-goals-model-explained/). 

The purpose of this work is to investigate xG, and some related statistical measures, to ascertain which could be used to predict a club's success in the Premier League (EPL). The figures looked at are as follows:

* Expected goals (xG): the sum of xG from all goalscoring opportunities in a season
* Expected goals against (xGA): the sum of xG that opponents accumulated from goalscoring opportunities against the club
* Goals for (GF): the number of goals a club scored in a season
* Goals against (GA): the number of goals a club conceded in a season

The success of a club is measured by the number of points they won over the course of a Premier League season. Each of the above metrics are checked for correlation against EPL points won. Data from [FBref](https://fbref.com/en/comps/9/Premier-League-Stats) covers seasons 2017/18, 2018/19, 2019/20, 2020/21. As 20 clubs compete each season, each scatterplot contains 80 data points. There are some limitations to this relatively small sample size, but this is unavoidable as xG data has only been available on FBref since the 2017/18 season.

The following questions are addressed:

1. Are expected goals for/against a better predictor of EPL position than actual goals for/against?
2. Do attack (xG/goals scored) or defence (xGA/goals conceded) statistics have differing levels of correlation?
3. Can we use xG, xGA, GF, or GA to predict a club's success from their siutation mid season?

### xG

```python
import matplotlib.pyplot as plt
from scipy import stats
import csv

x=[]
y=[]

with open('xG1718_2021.csv', 'r') as csvfile:
    plots= csv.reader(csvfile, delimiter=',')
    for row in plots:
        x.append(float(row[0]))
        y.append(float(row[1]))

slope, intercept, r, p, std_err = stats.linregress(x, y)

def myfunc(x):
  return slope * x + intercept

mymodel = list(map(myfunc, x))

plt.scatter(x, y)
plt.plot(x, mymodel)
plt.title("Relationship between xG and EPL Points: 2017/18 - 2020/21")
plt.xlabel("Points (1 season)")
plt.ylabel("xG (1 season)")
plt.show()

print("Correlation Coefficient:")
print(r)
```
